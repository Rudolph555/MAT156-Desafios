## PROBLEMA (Anterior Ejercicio)

Un ingeniero civil requiere 4800 m³ de arena, 5810 m³ de grano fino y 5690 m³ de grano grueso.

Las proporciones de estos materiales que se encuentran en tres canteras son las siguientes:

```python
import pandas as pd

# Crear una tabla para mostrar las proporciones de cada material en las canteras
data = {
    "Cantera": ["Cantera 1", "Cantera 2", "Cantera 3"],
    "Arena (%)": [52, 20, 25],
    "Grano Fino (%)": [30, 50, 20],
    "Grano Grueso (%)": [18, 30, 55]
}

# Convertir los datos en un DataFrame de pandas
df = pd.DataFrame(data)

# Mostrar la tabla
df
```

#### Objetivo:

Determinar la cantidad de metros cúbicos que se debe transportar desde cada cantera para cumplir con los requisitos.

### Resolviendo el sistema de ecuaciones por JACOBI:

```python
import numpy as np

# Definición de la matriz A y el vector b
A = np.array([[0.52, 0.2, 0.25],
              [0.3, 0.5, 0.2],
              [0.18, 0.3, 0.55]])

b = np.array([4800, 5810, 5690])

# Número de variables
n = len(b)

# Inicialización de las soluciones
x = np.zeros(n)
x_new = np.zeros(n)

# Número máximo de iteraciones
max_iter = 20

# Método de Jacobi
for iteration in range(max_iter):
    for i in range(n):
        # Suma de los términos
        sigma = sum(A[i][j] * x[j] for j in range(n) if j != i)
        # Actualización de la solución
        x_new[i] = (b[i] - sigma) / A[i][i]

    # Copiar los nuevos valores a la solución anterior
    x = x_new.copy()

    # Mostrar resultados en cada iteración
    print(f"Iteración {iteration + 1}: {x}")

print(f"Solución final después de {max_iter} iteraciones: {x}")
```
