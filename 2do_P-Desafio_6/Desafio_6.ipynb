##**DESAFIO 6 - SISTEMAS MAL CONDICIONADOS DE 3X3 CON GRAFICO**  
Univ. Ruddy Cruz


                  Ejemplo planteado:  
                  9x + 2y + 3z = 14  
                  10x + 2y + 3z = 15  
                  10x + 2.0001y + 3z = 15.0001
Este sistema está mal condicionado porque las filas de la matriz de coeficientes son casi linealmente dependientes. Se observa que la segunda y tercera ecuación son muy similares a la primera. Pequeños cambios en los coeficientes o en el término independiente pueden provocar grandes variaciones en la solución. La proximidad de las filas al ser linealmente dependientes resulta en un determinante cercano a cero, lo que indica un sistema inestable numéricamente.

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de coeficientes y vector de términos independientes
A = np.array([[9, 2, 3],
              [10, 2, 3],
              [10, 2.0001, 3]])
b = np.array([14, 15, 15.0001])

# a) Determinante
determinante = np.linalg.det(A)
print(f"Determinante: {determinante}")

# b) Matriz identidad
identidad = np.eye(3)
print(f"Matriz Identidad:\n{identidad}")

# c) Número de condición
condicion = np.linalg.cond(A)
print(f"Número de condición: {condicion}")

# d) Pequeños cambios y cambio en la solución
# Perturbación en los coeficientes
A_perturbado = np.array([[9, 2, 3],
                         [10, 2.0001, 3],
                         [10, 2.0001, 3.0001]])
b_perturbado = np.array([14, 15.0001, 15.0002])

# Solución original
try:
  solucion_original = np.linalg.solve(A, b)
  print(f"Solución original: {solucion_original}")
except np.linalg.LinAlgError:
  print("Solución original: El sistema es singular o casi singular y no tiene solución única.")

# Solución perturbada
try:
  solucion_perturbada = np.linalg.solve(A_perturbado, b_perturbado)
  print(f"Solución perturbada: {solucion_perturbada}")
except np.linalg.LinAlgError:
  print("Solución perturbada: El sistema es singular o casi singular y no tiene solución única.")


# Gráfico (visualización simplificada debido a la dificultad de representar planos casi coincidentes)
# En este caso, la gráfica 3D no es tan informativa porque los planos son casi coincidentes.
# Se muestra una gráfica simple de los valores de x,y,z para las soluciones si existen.

if 'solucion_original' in locals():
  plt.figure(figsize=(6,4))
  plt.plot(solucion_original[0], solucion_original[1], solucion_original[2], 'ro', label='Solución Original')
  if 'solucion_perturbada' in locals():
    plt.plot(solucion_perturbada[0], solucion_perturbada[1], solucion_perturbada[2], 'go', label='Solución Perturbada')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Solución del Sistema (si existe)')
  plt.legend()
  plt.show()
else:
  print("No se pudo graficar ya que el sistema no tiene solución única.")
```
