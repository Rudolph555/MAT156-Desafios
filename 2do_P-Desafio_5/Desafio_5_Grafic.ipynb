```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz A y el vector b
A = np.array([[1.0001, 1.0000],
              [1.0000, 1.0000]])
b = np.array([2, 2])

# Resolver el sistema Ax = b
x = np.linalg.solve(A, b)

# Crear un rango de valores para x
x_values = np.linspace(0, 3, 100)

# Ecuaciones del sistema
y1 = (2 - 1.0001 * x_values) / 1.0000  # de la primera ecuación
y2 = (2 - 1.0000 * x_values) / 1.0000  # de la segunda ecuación

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x_values, y1, label='1.0001x + 1.0000y = 2', color='blue')
plt.plot(x_values, y2, label='1.0000x + 1.0000y = 2', color='orange')
plt.scatter(x[0], x[1], color='red', label='Solución (x, y)')
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.title('Gráfica del sistema de ecuaciones')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
